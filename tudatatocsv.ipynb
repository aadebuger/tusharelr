```python
pro = ts.pro_api()
df = pro.daily(ts_code='600115.SH', start_date='20160701', end_date='20180718')
```

```python
print(df)
def convert_to_sp(item):
    return item[6:8]+'-'+ item[4:6]+'-'+item[2:4]
df['Date'] = df['trade_date'].apply(convert_to_sp)
df = df.rename(columns={"vol":"Volume","close":"Close","low":"Low","high":"High","open":"Open"})
df = df.iloc[::-1]
df.to_csv("600115.csv",columns=['Date','Open','High','Low','Close','Volume'],index=False)
```
